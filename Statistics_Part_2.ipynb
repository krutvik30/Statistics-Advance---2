# **Theory Questions**

1. What is hypothesis testing in statistics?


=> It is a method to make inferences about a population using sample data.



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?


=> The null hypothesis states there is no effect; the alternative states there is an effect.



3. What is the significance level in hypothesis testing, and why is it important?


=> It is the probability of rejecting a true null hypothesis, used to control Type 1 error.



4. What does a P-value represent in hypothesis testing?


=> It is the probability of obtaining test results at least as extreme as the observed results, under the null hypothesis.



5. How do you interpret the P-value in hypothesis testing?


=> If the P-value ≤ significance level, reject the null hypothesis.



6. What are Type 1 and Type 2 errors in hypothesis testing?


=> Type 1 error: rejecting a true null hypothesis. Type 2 error: failing to reject a false null hypothesis.



7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?


=> A one-tailed test checks deviation in one direction; a two-tailed test checks both.



8. What is the Z-test, and when is it used in hypothesis testing?


=> Z-test is used when population variance is known and sample size is large.



9. How do you calculate the Z-score, and what does it represent in hypothesis testing?


=> Z=(X-μ)/σ; it represents how many standard deviations a value is from the mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?


=> Used when sample size is small and population variance is unknown.



11. What is the difference between a Z-test and a T-test?


=> Z-test is used with known variance; T-test is used with unknown variance.



12. What is the T-test, and how is it used in hypothesis testing?


=> T-test compares sample mean(s) with a population or another sample to test hypotheses.



13. What is the relationship between Z-test and T-test in hypothesis testing?


=> Both test means; T-test is a generalization of the Z-test for small samples or unknown variance.



14. What is a confidence interval, and how is it used to interpret statistical results?


=> It is a range where the population parameter is expected to lie with a certain confidence level.



15. What is the margin of error, and how does it affect the confidence interval?


=> It is half the width of the confidence interval; larger margin increases the interval range.



16. How is Bayes' Theorem used in statistics, and what is its significance?


=> It updates the probability of a hypothesis based on new evidence.



17. What is the Chi-square distribution, and when is it used?


=> It is used in tests of independence and goodness-of-fit for categorical data.



18. What is the Chi-square goodness of fit test, and how is it applied?


=> It compares observed frequencies with expected ones to test distribution fit.



19. What is the F-distribution, and when is it used in hypothesis testing?


=> Used to compare variances and in ANOVA tests.



20. What is an ANOVA test, and what are its assumptions?


=> It tests if means of three or more groups are equal. Assumptions: normality, homogeneity of variances, independence.



21. What are the different types of ANOVA tests?


=> One-way ANOVA, Two-way ANOVA, and Repeated Measures ANOVA.



22. What is the F-test, and how does it relate to hypothesis testing?


=> It tests the ratio of variances to assess the equality of variances between groups.

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(alpha=0.05, beta=0.2, mu0=0, mu1=1, sigma=1):

    x = np.linspace(-4, 6, 1000)

    y0 = norm.pdf(x, mu0, sigma)

    y1 = norm.pdf(x, mu1, sigma)

    z_critical = norm.ppf(1 - alpha, mu0, sigma)

    plt.plot(x, y0, label='H0: N(0,1)', color='blue')
    plt.plot(x, y1, label='H1: N(1,1)', color='red')

    x_alpha = np.linspace(z_critical, 6, 100)
    plt.fill_between(x_alpha, norm.pdf(x_alpha, mu0, sigma), color='blue', alpha=0.3, label='Type I Error (α)')

    x_beta = np.linspace(-4, z_critical, 100)
    plt.fill_between(x_beta, norm.pdf(x_beta, mu1, sigma), color='red', alpha=0.3, label='Type II Error (β)')

    plt.axvline(z_critical, color='black', linestyle='--', label=f'Critical value (z={z_critical:.2f})')
    plt.title('Type I and Type II Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_type1_type2()
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=30)
population_mean = 100

# Z-test
z_score = (np.mean(data) - population_mean) / (15 / np.sqrt(len(data)))  # using known population std dev
z_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# T-test
t_statistic, t_p_value = stats.ttest_1samp(data, population_mean)

print("Z-test:")
print(f"Z-score = {z_score:.4f}, P-value = {z_p_value:.4f}")
print("\nT-test:")
print(f"T-statistic = {t_statistic:.4f}, P-value = {t_p_value:.4f}")
```
